## 7)Test the Analysis of Variance between etest_p and mba_p at signifance level 5%.(Make decision using Hypothesis Testing)

Null Hypothesis (𝐻0): There is no significant difference between etest_p and mba_p.

Alternative Hypothesis (H1): There is a significant difference between etest_p and mba_p.

```python
from scipy import stats

# Perform ANOVA
f_statistic, p_value = stats.f_oneway(df['etest_p'], df['mba_p'])

print(f'F-Statistic: {f_statistic}')
print(f'P-Value: {p_value}')

# Significance Level
alpha = 0.05

# Decision
if p_value < alpha:
    print("Reject the Null Hypothesis - Significant Difference Exists")
else:
    print("Fail to Reject the Null Hypothesis - No Significant Difference.")
```

### The p-value ≤ 0.05, it means the difference is statistically significant, and we reject the null hypothesis.

## 8)Test the similarity between the degree_t(Sci&Tech) and specialisation(Mkt&HR) with respect to salary  at significance level of 5%.(Make decision using Hypothesis Testing)

Null Hypothesis (H0): There is no significant difference in the salaries between degree_t (Sci&Tech) and specialisation (Mkt&HR).

Alternative Hypothesis(H1): There is a significant difference in the salaries between these two groups.

```python
# Filter the data for the specific groups
sci_tech_salary = df[df['degree_t'] == 'Sci&Tech']['salary'].dropna()
mkt_hr_salary = df[df['specialisation'] == 'Mkt&HR']['salary'].dropna()

print(f'Sci&Tech Salary Count: {len(sci_tech_salary)}')
print(f'Mkt&HR Salary Count: {len(mkt_hr_salary)}')
```

```python
from scipy import stats

# Perform Independent T-Test
t_statistic, p_value = stats.ttest_ind(sci_tech_salary, mkt_hr_salary, equal_var=False)

print(f'T-Statistic: {t_statistic}')
print(f'P-Value: {p_value}')

# Significance Level
alpha = 0.05

# Decision
if p_value < alpha:
    print("Reject the Null Hypothesis - Significant Difference in Salaries")
else:
    print("Fail to Reject the Null Hypothesis - No Significant Difference in Salaries")
```

## 10)What is the probability Density Function of the salary range from 700000 to 900000?

```python
from scipy.stats import norm

# Calculate mean and standard deviation of salary
mean_salary = df['salary'].mean()
std_salary = df['salary'].std()

# Calculate Z-Scores
z_700000 = (700000 - mean_salary) / std_salary
z_900000 = (900000 - mean_salary) / std_salary

# Calculate CDF values using the standard normal distribution
p_700000 = norm.cdf(z_700000)
p_900000 = norm.cdf(z_900000)

# Calculate the probability of salary between 700000 and 900000
probability = p_900000 - p_700000
print(f"Probability that salary is between 700,000 and 900,000: {probability:.4f}")
```

### A probability of 0.0006 indicates that only 0.06% of the population in the dataset has a salary in this range.

## 11)Test the similarity between the degree_t(Sci&Tech)with respect to etest_p and mba_p at significance level of 5%.(Make decision using Hypothesis Testing)

Null Hypothesis (H0): There is no significant difference in etest_p and mba_p for students with a Sci&Tech degree.

Alternative Hypothesis(H1): There is a significant difference in etest_p and mba_p for students with a Sci&Tech degree


```python
# Filter the data for the specific groups
sci_tech_etest = df[df['degree_t'] == 'Sci&Tech']['etest_p'].dropna()
sci_tech_mba = df[df['degree_t'] == 'Sci&Tech']['mba_p'].dropna()

print(f'Sci&Tech_etest: {len(sci_tech_etest)}')
print(f'Sci&Tech_mba: {len(sci_tech_mba)}')
```

```python
from scipy import stats

# Perform paired T-Test

t_stat, p_value = stats.ttest_rel(sci_tech_etest, sci_tech_mba)

print(f"Paired T-Test: t-stat={t_stat}, p-value={p_value}")

# Significance Level
alpha = 0.05

if p_value < alpha:
    print("Reject the Null Hypothesis - Significant Difference in Scores (etest_p vs mba_p)")
else:
    print("Fail to Reject the Null Hypothesis - No Significant Difference in Scores (etest_p vs mba_p)")
```

### 14. Perform Chi-Square test.Let's say you want to check if there is a relationship between degree_t and status (placed or not).

Null Hypothesis (𝐻0): There is no significant difference between degree type and status.

Alternative Hypothesis (H1): There is a significant difference between degree type and status.

```python
from scipy.stats import chi2_contingency

# Create a contingency table
contingency_table = pd.crosstab(df['degree_t'], df['status'])
print(contingency_table)
```

```python
# Perform Chi-Square Test
chi2_stat, p_value, dof, expected = chi2_contingency(contingency_table)

print(f"Chi-Square Statistic: {chi2_stat}")
print(f"P-Value: {p_value}")
print(f"Degrees of Freedom: {dof}")
print(f"Expected Frequencies: \n{expected}")

# Significance Level
alpha = 0.05

# Decision
if p_value < alpha:
    print("Reject the Null Hypothesis")
else:
    print("Fail to Reject the Null Hypothesis")
```
