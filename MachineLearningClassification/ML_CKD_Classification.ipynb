```python
dataset=pd.get_dummies(df,drop_first=True,dtype="int")
```

```python
X= dataset.iloc[:,:-1].to_numpy()
y= dataset.iloc[:,-1].to_numpy()
```

```python
X.shape
```

```python
# Split data into training and testing sets
from sklearn.model_selection import train_test_split
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.2, random_state=0)
```

```python
# Performing standard scaling
from sklearn.preprocessing import StandardScaler
sc = StandardScaler()
X_train = sc.fit_transform(X_train)
X_test = sc.transform(X_test)
```

#### Naive Bayes (probabilistic classification algorithm) : Naive Bayes is mainly used for classification problems, not for regression. It works on the principle of Bayes' Theorem and assumes that features are independent (which is often not true in real-world scenarios, but it still performs well).

#### For Binary Classification: Use BernoulliNB or GaussianNB depending on whether your data is binary or continuous.

#### For Multiclass Classification: Use MultinomialNB if working with text or categorical data.

#### For Continuous Data in Multiclass: Use GaussianNB.

```python
import pickle
from sklearn.model_selection import GridSearchCV
from sklearn.metrics import accuracy_score, precision_score, recall_score, roc_auc_score,f1_score
from sklearn.linear_model import LogisticRegression
from sklearn.tree import DecisionTreeClassifier
from sklearn.ensemble import RandomForestClassifier, GradientBoostingClassifier
from sklearn.svm import SVC
from sklearn.neighbors import KNeighborsClassifier
from xgboost import XGBClassifier
from sklearn.naive_bayes import BernoulliNB

# Define algorithms and parameter grids
algorithms = {
    'LogisticRegression': {
        'model': LogisticRegression(),
        'params': {
            'C': [0.01, 0.1, 1, 10],
            'penalty' : ['l1', 'l2', 'elasticnet'],
            'solver': ['liblinear', 'lbfgs'],
            'max_iter': [100, 200]
        }
    },
   'SVM': {
        'model': SVC(),
        'params': {
            'C': [0.1, 1, 10],
            'kernel': ['linear', 'rbf', 'poly'],
            'gamma': ['scale', 'auto']
        }
    },
    'DecisionTree': {
        'model': DecisionTreeClassifier(),
        'params': {
            'criterion': ['gini', 'entropy','log_loss'],
            'splitter':['best', 'random'],
            'max_depth': [5, 10, 20],
            'min_samples_split': [2, 5, 10],
            'min_samples_leaf': [1, 5, 10]
        }
    },
    'RandomForest': {
        'model': RandomForestClassifier(),
        'params': {
            'n_estimators': [100, 150, 200],
            'max_depth': [10, 20, None],
            'criterion': ['gini', 'entropy','log_loss'],
            'max_features': ['sqrt', 'log2'],
            'ccp_alpha': [0.0, 0.01, 0.1, 1.0]
        }
    },  
    'GradientBoosting': {
        'model': GradientBoostingClassifier(),
        'params': {
            'n_estimators': [100, 150, 200],
            'learning_rate': [0.01, 0.1, 1],
            'max_depth': [3, 5, 7],
            'min_samples_split': [2, 5, 10]
        }
    },
    'XGBoost': {
        'model': XGBClassifier(),
        'params': {
            'n_estimators': [100, 150, 200],
            'learning_rate': [0.01, 0.1, 1],
            'max_depth': [3, 5, 7],
            'min_child_weight': [1, 3, 5]
        }
    },
    'KNN': {
        'model': KNeighborsClassifier(),
        'params': {
            'n_neighbors': [3, 5, 7, 9],  # sqrt(399)= 19.97 ==> so we need to consider 19 or 21 as n_neighbors.399 is total number of rows in the dataset.
            'weights': ['uniform', 'distance'],
            'algorithm': ['ball_tree', 'kd_tree', 'brute'],
            'p': [1, 2]
        }
    },
    'BernoulliNB': {
        'model': BernoulliNB(),
        'params': {
            'alpha': [0.1, 0.5, 1.0],
            'binarize': [0.0, 0.5, 1.0]
        }
    }
}
```

```python

# Initialize results dictionary to store model performance
results = {}
start = time.time()

# Perform Grid Search for each algorithm
for algo_name, config in algorithms.items():
    print(f"Training {algo_name}...")
    grid = GridSearchCV(
        estimator=config['model'],
        param_grid=config['params'],
        scoring='accuracy',
        refit=True,
        cv=4,
        verbose=1,
        n_jobs=-1
    )
    grid.fit(X_train, y_train)
    
    # Predict on test data
    y_pred = grid.predict(X_test)
    
    # Calculate metrics
    accuracy = accuracy_score(y_test, y_pred)
    precision = precision_score(y_test, y_pred)
    recall = recall_score(y_test, y_pred)
    roc_auc = roc_auc_score(y_test, y_pred) # Use ROC-AUC if you care about overall model performance across all classification thresholds.
    f1_weighted = f1_score(y_test, y_pred, average='weighted') #Use F1 Score if you care about both false positives and false negatives.
    
    # Store results
    results[algo_name] = {
        'best_params': grid.best_params_,
        'accuracy': accuracy,
        'f1weighted' : f1_weighted,
        'precision': precision,
        'recall': recall,
        'roc_auc': roc_auc,
        'best_model': grid.best_estimator_
    }

end = time.time()
print('Time taken in grid search: %0.2f seconds'%(end-start))
```

```python
# Print Results
for algo_name, result in results.items():
    print(f"\n{algo_name}:")
    print(f"Best Parameters: {result['best_params']}")
    print(f"Accuracy: {result['accuracy']}")
    print(f"f1weighted: {result['f1weighted']}")
    print(f"precision: {result['precision']}")
    print(f"Recall: {result['recall']}")
    print(f"Roc_auc: {result['roc_auc']}")
```

```python
# Find best algorithm based on accuracy
best_algo_name = max(results, key=lambda k: results[k]['accuracy']) # As class is not imbalanced so i'm using accuracy to check the results.
best_model = results[best_algo_name]['best_model']
```

```python
# Save the best model using pickle
filename = "CKD_best_model.pkl"
with open( filename, 'wb') as file:
    pickle.dump(best_model, file)
print(f"Best model ({best_algo_name}) saved to {filename}")
```

```python
# Load and Predict using the saved model
with open('CKD_best_model.pkl', 'rb') as file:
    loaded_model = pickle.load(file)
predictions = loaded_model.predict(X_test)
print("Predictions using loaded model:", predictions)
```
